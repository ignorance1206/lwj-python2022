## 把Day这一列1990/12/19这种格式的数据转换成日期格式

```python
# format 形式一定要对应原始数据的样子 1990/12/19 %Y/%m/%d 1990-12-19 %Y-%m-%d
# %h %m %s
# 试试 %y/%m/%d
data['Day'] = pd.to_datetime(data['Day'],format = '%Y/%m/%d')
data
```
